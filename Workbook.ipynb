```python
# Use tf-idf features for NMF.
print("Extracting count vectoriser features for NMF...")
count_vectorizer = CountVectorizer(max_df=0.95, min_df=2,
                                   stop_words='english')
t0 = time()
count_vec = count_vectorizer.fit_transform(df_train['content'])
print("done in %0.3fs." % (time() - t0))
```

```python
count_vectorizer.vocabulary_
```

```python
n_components = 100
nmf = NMF(n_components=n_components, random_state=1,
          alpha=.1, l1_ratio=.5)
W_train = nmf.fit_transform(count_vec)
H_train = nmf.components_
```

```python
lst = []
cols = []
for i in range(1,101):
    cols.append("val_" + str(i))
for index, row in df_train.iterrows():
    tmp = [row['title']]
    for i in range(100):
        tmp.append(W_train[index][i])
    lst.append(tmp)
```

```python
df_train_embedding = pd.DataFrame(lst, columns=['title'] + cols)
```

```python
df_train_embedding.head()
```

```python
artists = ["Leonardo DiCaprio","Amitabh Bachchan","Gal Gadot","Ed Sheeran"]
artists = [ppl.lower() for ppl in artists]

sportsPerson =  ["Andre Agassi","Maria Sharapova","Sachin Tendulkar","Lionel Messi"]
sportsPerson = [ppl.lower() for ppl in sportsPerson]

politicians = ["Barack Obama","Narendra Modi","Vladimir Putin"]
politicians = [ppl.lower() for ppl in politicians]

scientists = ["Albert Einstein","Isaac Newton","Dmitri Mendeleev","Charles Darwin"]
scientists = [ppl.lower() for ppl in scientists]

influencers = ["Steve Jobs","Larry Page",
             "Bill Gates","Alan Greenspan","Sal Khan","Geoffrey Hinton"]
influencers = [ppl.lower() for ppl in influencers]
```

```python
embedding_artists = df_train_embedding[df_train_embedding['title'].isin(artists)].mean()
embedding_sportsPerson = df_train_embedding[df_train_embedding['title'].isin(sportsPerson)].mean()
embedding_politicians = df_train_embedding[df_train_embedding['title'].isin(politicians)].mean()
embedding_scientists = df_train_embedding[df_train_embedding['title'].isin(scientists)].mean()
embedding_influencers = df_train_embedding[df_train_embedding['title'].isin(influencers)].mean()
```

```python
W.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
for i in range(25):
    print(df_test['title'][i],round(cosine_similarity([W[i]], [embedding_artists])[0][0],2),
          round(cosine_similarity([W[i]], [embedding_sportsPerson])[0][0],2),
          round(cosine_similarity([W[i]], [embedding_politicians])[0][0],2),
          round(cosine_similarity([W[i]], [embedding_scientists])[0][0],2),
          round(cosine_similarity([W[i]], [embedding_influencers])[0][0],2))
```

```python
lst = []
for i in range(25):
    tmp = [df_test['title'][i]]
    v1 = cosine_similarity([W[i]], [embedding_artists])[0][0]
    v2 = cosine_similarity([W[i]], [embedding_sportsPerson])[0][0]
    v3 = cosine_similarity([W[i]], [embedding_politicians])[0][0]
    v4 = cosine_similarity([W[i]], [embedding_scientists])[0][0]
    v5 = cosine_similarity([W[i]], [embedding_influencers])[0][0]
    msum = v1 + v2 + v3 + v4 + v5
    tmp = tmp + [v1/msum,v2/msum,v3/msum,v4/msum,v5/msum]
    lst.append(tmp)
    
    
    
    
    
```

```python
df_test_prediction = pd.DataFrame(lst, columns=['title','Artist','SportsPerson','Politician/Leader','Scientist/Researcher','Influencer'])
```
